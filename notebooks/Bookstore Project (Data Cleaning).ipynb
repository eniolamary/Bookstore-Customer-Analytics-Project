# Data Cleaning
## 1. Standardize text columns

### Why this comes first:
Raw categorical data often contains inconsistencies (e.g., different casing, spelling variations, and duplicate labels). Standardizing ensures that all categories are uniform before analysis or grouping operations.

### What was done:
- Converted text values to consistent formatting
- Unified duplicate category labels (e.g., “False”, “No”, “0”)
- Replaced inconsistent entries with standard labels

### Code:

```python
string_cols = df.select_dtypes(include=['object', 'string']).columns
df[string_cols] = df[string_cols].apply(lambda x: x.str.strip().str.lower().str.title())
```

```python
columns = [
    'region',
    'home_store',
    'loyalty_member',
    'has_escalation',
    'escalation_reason'
]
for col in columns:
    print(f"\n{'='*60}")
    print(df[col].value_counts(dropna=False))
```

```python
df['home_store'] = df['home_store'].replace({'Yyc Dt' : 'YYC-DT', 'Yyc-Dt' : 'YYC-DT', 
                                             'Yyc Nw' : 'YYC-NW', 'Yyc-Nw' : 'YYC-NW', 
                                             'Yyc Se' : 'YYC-SE', 'Yyc-Se' : 'YYC-SE'})
df['region'] = df['region'].replace({'Calg Downtown' : 'Calgary Downtown', 
                                     'Calg Nw' : 'Calgary Nw', 'Calg Se' : 'Calgary Se'})
df['loyalty_member'] = df['loyalty_member'].replace({'Y': 'Yes', 'True': 'Yes', '1' : 'Yes', 
                                                     'N': 'No', 'False': 'No', '0' : 'No'})
df['has_escalation'] = df['has_escalation'].replace({'Y': 'Yes', 'True': 'Yes', '1' : 'Yes', 
                                                     'N': 'No', 'False': 'No', '0' : 'No'})
df['escalation_reason'] = df['escalation_reason'].replace({'Websiteissue' : 'Website Issue', 'Latedelivery' : 'Late Delivery', 
                                                           'Damagedbook' : 'Damaged Book', 'Rudestaff' : 'Rude Staff', 
                                                           'Billingerror' : 'Billing Error'})
```

### Outcome:
Several categorical variables contained inconsistent labels representing the same values. For example, the home_store column included variations such as "Yyc Dt" and "Yyc-Dt", while loyalty_member and has_escalation contained multiple representations of the same response (e.g., "Y", "True", and "1" all indicated "Yes").

These inconsistencies were standardized into a single, consistent format for each category. This cleaning step improves data quality by ensuring that identical values are not treated as separate categories during analysis. It also produces more accurate frequency counts, clearer visualizations, and more reliable statistical analyses and machine learning models.

## 2. Fix Missing values (NA/nulls)

### Why:

Missing values can distort analysis and machine learning models. The treatment depends on the type of data and its business meaning.

### What was done:
- Numerical columns → imputed using median
- Categorical columns → filled with "Unknown"
- Business-rule columns → handled using conditional logic


```python
df.isnull().sum()
```

```python
# Numerical columns (median imputation)
numeric_cols = [
    'annual_visits',
    'online_ratio',
    'total_purchases',
    'total_spent',
    'avg_purchase_value',
    'visits_2023',
    'customer_rating'
]
for col in numeric_cols:
    df[col] = (
        df.groupby('region')[col]
          .transform(lambda x: x.fillna(x.median()))
    )
# Categorical columns
categorical_cols = [
    'loyalty_member',
    'has_escalation'
]
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')

# Escalation reason
df.loc[df['has_escalation'] == 'No', 'escalation_reason'] = 'No Escalation'
df.loc[
    (df['has_escalation'] == 'Yes') & 
    (df['escalation_reason'].isna()),
    'escalation_reason'
] = 'Unknown'

df.loc[df['has_escalation'] == 'Unknown', 'escalation_reason'] = 'Unknown'
```

```python
df.isnull().sum()
```

### Outcome:

Missing values were handled using different imputation strategies depending on the type and context of each variable.

For the numerical variables, missing values were replaced with the median of their respective region rather than the overall dataset median. This approach preserves regional differences in customer behavior while reducing the influence of extreme values, making it a more robust choice for skewed numerical data.

For the binary categorical variables (loyalty_member and has_escalation), missing values were replaced with "Unknown" instead of assuming a "Yes" or "No" response. This preserves the uncertainty in the data and avoids introducing bias through incorrect assumptions.

The escalation_reason column was then updated based on the values in has_escalation:
Customers with no escalation were assigned "No Escalation" since a reason is not applicable.
Customers with an escalation but a missing reason were assigned "Unknown" to indicate that the escalation occurred, but its reason was unavailable.
Customers whose escalation status was also unknown were assigned "Unknown" for the escalation reason to maintain consistency between the two related variables.

By applying context-specific imputation methods, the dataset retains as much information as possible while ensuring logical consistency between related features.

### Note on Missing Date Values

Although missing values in the numerical and categorical variables were addressed, the missing values in the date columns (signup_date, first_purchase_date, and last_purchase_date) were intentionally retained.

Unlike numerical or categorical features, these dates could not be reliably inferred without making unsupported assumptions. Imputing arbitrary dates could introduce misleading information and compromise the accuracy of any time-based analyses. Therefore, the missing dates were preserved as NaT (Not a Time), accurately reflecting that the information was unavailable rather than artificially generated.

When date-derived features such as customer tenure and purchase lifespan were later created for machine learning, the resulting numerical missing values were handled separately using an appropriate imputation strategy.

## 3. Fix Incorrect data types

### Why:
Correct data types are essential for analysis (especially dates and numeric calculations).

### What was done:
- Converted ID fields to string type
- Converted date columns to datetime format

```python
# Date columns
date_cols = [
    'signup_date',
    'first_purchase_date',
    'last_purchase_date'
]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='mixed', errors='coerce')

df['customer_id'] = df['customer_id'].astype(str)
```

### Outcome:
The data types of selected columns were corrected to ensure they accurately represented the information they contained.

The signup_date, first_purchase_date, and last_purchase_date columns were converted from text to the datetime data type. This enables date-based operations such as calculating customer tenure, purchase lifespan, extracting date components (e.g., year or month), and performing time-series analysis. Using errors='coerce' ensures that any invalid or improperly formatted dates are safely converted to missing values (NaT) rather than causing errors during processing.

Additionally, the customer_id column was converted to a string because it serves as a unique identifier rather than a numerical measurement. Treating it as text prevents it from being mistakenly included in numerical calculations, statistical summaries, or machine learning models where it has no predictive value.
Correcting these data types ensures that each variable is stored in the most appropriate format, improving data integrity and enabling accurate analysis in subsequent stages of the project.

## 4. Remove duplicates

### Why:

Duplicate records can bias analysis and inflate metrics like revenue or customer counts.

### What was done:

- Identified duplicate rows
- Removed duplicates while preserving first occurrence

```python
df.drop_duplicates(inplace=True)
```

### Outcome:
Duplicate records were identified and removed to ensure that each row in the dataset represented a unique customer observation. Duplicate entries can distort summary statistics, bias exploratory data analysis, and negatively impact machine learning models by giving certain observations more influence than they should have.
By removing duplicate rows, the dataset becomes more reliable and accurately reflects the underlying customer data, leading to more trustworthy analyses and model predictions.
